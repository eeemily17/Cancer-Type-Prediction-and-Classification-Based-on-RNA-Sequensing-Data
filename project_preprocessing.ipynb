```python
# load the raw dataset
import numpy as np
import pandas as pd
from pandas import DataFrame

rawdf = pd.read_csv("unc.edu_PANCAN_IlluminaHiSeq_RNASeqV2.geneExp.tsv", sep="\t", index_col=0)
rawdf.head()
```

```python
# transpose raw dataset so row is patient samples and column is list of genes
processeddf = rawdf.transpose()
processeddf.head()
```

```python
# check to see if there is any missing values
processeddf.isnull().any().any()
# answer is no
```

```python
# check to see if any duplicate rows/samples
processeddf.index.get_duplicates()
```

```python
# check to see if any duplicate columns/features
processeddf.transpose().index.get_duplicates()
```

```python
# check to see if there is any feature (i.e. column) has all zero values so we will delete them
removedAllZeroColdf = processeddf.loc[:, (processeddf != 0).any(axis=0)]
removedAllZeroColdf.shape
```

```python
# calculate how many features/columns has all zeros and we have removed them
removedFeatures = processeddf.shape[1] - removedAllZeroColdf.shape[1]
removedFeatures
# 212 features are removed from the data set
```

```python
# list out the removed 212 features and manually confirm that they are right
original = DataFrame(processeddf.columns)
original.columns = ['genelist']
removed = DataFrame(removedAllZeroColdf.columns)
removed.columns = ['genelist']

original[~original.genelist.isin(removed.genelist)].head()
```

```python
# manually confirm the first example gene #7661 (HBII-52-24|100036563) is all empty so it is removed

count = 0
potentiallyAllZeroColumn = processeddf['HBII-52-24|100036563']
for i in range(10471):
    if(potentiallyAllZeroColumn[i] == 0.0):
        count = count+1
        
print(count)
```
